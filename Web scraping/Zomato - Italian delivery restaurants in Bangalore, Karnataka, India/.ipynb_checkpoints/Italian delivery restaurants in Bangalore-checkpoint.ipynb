```python
# import the necessary libraries

import pandas as pd
import numpy as np
from selenium import webdriver
from selenium.webdriver.common.keys import Keys

from warnings import filterwarnings
filterwarnings('ignore')
```

```python
# specify the location of the chrome driver
driver1 = webdriver.Chrome('C:\\Users\\kvham\\Desktop\\SHEEREEN\\Data Science projects\\Web scraping\\Zomato - Italian delivery restaurants in Bangalore, Karnataka, India\\chromedriver_win32\\chromedriver')
driver2 = webdriver.Chrome('C:\\Users\\kvham\\Desktop\\SHEEREEN\\Data Science projects\\Web scraping\\Zomato - Italian delivery restaurants in Bangalore, Karnataka, India\\chromedriver_win32\\chromedriver')
```

```python
# target url to scrap
# zomato.com + bangalore city + italian restaurants
url = 'https://www.zomato.com/bangalore/restaurants/italian?category=1'
url
```

```python
driver1.get(url)
data = driver1.find_elements_by_class_name('jumbo-tracker')
len(data)
```

```python
restaurants = []

# loop through each of the restaurant details
for details in data:
    restaurant = {}
    restaurant['name'] = details.find_element_by_tag_name('h4').text
    restaurant['time_to_deliver'] = details.find_element_by_tag_name('a').find_elements_by_tag_name('p')[-1].text.split()[0]
    restaurant['offer'] = details.find_element_by_tag_name('a').find_elements_by_tag_name('p')[0].text \
                            if len(details.find_element_by_tag_name('a').find_elements_by_tag_name('p')) > 1 \
                            else None
    restaurant['items'] = details.find_elements_by_tag_name('a')[-1].find_elements_by_xpath('//div[2]//p')[2].text
    restaurant['average_cost'] = details.find_elements_by_tag_name('a')[-1].find_element_by_xpath('//div[2]//p[2]').text
    
    restaurant_url = details.find_element_by_tag_name('a').get_attribute('href')
    
    restaurant_data = driver2.get(restaurant_url)
    
    restaurant_info = driver2.find_element_by_xpath('//main//div//section[3]')
    ratings_and_reviews = restaurant_info.find_element_by_tag_name('h1').find_element_by_xpath('./..//section'). \
                                                                         text.split('\n')
    restaurant['dining_rating'], restaurant['dining_no_reviews'] = ratings_and_reviews[0], ratings_and_reviews[1]
    restaurant['delivery_rating'], restaurant['delivery_no_reviews'] = ratings_and_reviews[3], ratings_and_reviews[4]
    restaurant['address'] = ', '.join(restaurant_info.find_element_by_xpath('//div//section'). \
                                                                 text.split('\n/\n')[-3:-6:-1])
    
    print(restaurant)
    restaurants.append(restaurant)
    
    
```

```python
# convert to dataframe
restaurants_df = pd.DataFrame.from_dict(restaurants)
restaurants_df
```

```python
# save dataframe as csv file
restaurants_df.to_csv('Italian_restaurants_in_bangalore.csv', mode='a', index=False)
```
